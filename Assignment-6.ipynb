```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('weather.csv')
```

```python
df
```

# Scatterplot

```python
sns.relplot(x="Date",y="MaxTemp",data=df)
```

        Scatter Plot between Date and MaxTemp

```python
sns.relplot(x="Date",y="MaxTemp",hue="MaxTemp",data=df)
```

```python
df.columns
```

```python
sns.relplot(x="Date",y="MinTemp",hue="RainToday",data=df)
```

        Scatter plot of Date and Temp with RainToday

```python
sns.relplot(x="Date",y="MaxTemp",hue="MaxTemp",style="RainTomorrow",data=df)
```


# Line graph

```python
sns.relplot(x="Date",y="MaxTemp",kind="line",data=df)

# with kind we mentioned that we want line graph
```

```python
sns.relplot(x="Date",y="Rainfall",kind="line",data=df)
```

    Plot between date and Rainfall

```python
sns.relplot(x="Date",y="MaxTemp",kind="line",hue="RainTomorrow",data=df)
```

```python
sns.relplot(x="Date",y="MaxTemp",kind="line",col="WindDir9am",col_wrap=4,height=3,data=df)

#col command divide the line graph on the basis of the column mentioned
# with the help of col_wrap we decide no of graph in one row
```

# Categorial Scatterplot

```python
sns.catplot(x="WindGustDir",y="MaxTemp",data=df)
```

```python
sns.catplot(y="MaxTemp",x="RainTomorrow",kind="swarm",data=df)
```

                     Swarm Plot

```python
sns.catplot(x="MaxTemp",y="RainTomorrow",kind="swarm",hue="RainToday",data=df)
```

# Box Plot

```python
sns.catplot(x="WindGustDir",y="MaxTemp",data=df,kind="box")
```

```python
sns.catplot(x='Cloud9am', y='MaxTemp', data=df,hue="RainToday",kind="box")
```

```python
sns.catplot(x='Cloud9am', y='MaxTemp', data=df,hue="RainToday",kind="boxen")
```

                  Boxen Plot
                  


# Bar Plot

```python
sns.catplot(x="WindSpeed9am",y="Humidity9am",data=df.iloc[0:50],hue="RainToday",kind="bar")
```

    Plot between Humidity and WindSpeed on RainyDay

```python
sns.catplot(y="Date",x="MinTemp",kind="bar",data=df[0:10],hue="RainToday")
```

    MinTemp with Rain Possibility


# Count Plot 

```python
sns.countplot(x=df["RainToday"],data=df)
```


# Histogram and KDE

```python
sns.distplot(df["Humidity9am"])
```

```python
sns.distplot(df["MaxTemp"],hist=False,rug=True)
```


# Joint Plot

```python
sns.jointplot(x="Temp9am",y="Humidity9am",data=df)
```

```python
sns.jointplot(x="Temp9am",y="Humidity9am",data=df,kind="hex")
```

```python
sns.jointplot(x="Temp9am",y="Humidity9am",data=df,kind="kde")
```


# Pair Plot

```python
sns.pairplot(data=df[["Temp9am","Humidity9am","Pressure9am","Cloud9am","RainToday"]],hue="RainToday")
```


# Pie Chart

```python
plt.figure(figsize=(15,8))
df['WindGustDir'].value_counts().plot(kind='pie')
```
